# TSNE

```python
from rdkit import Chem, DataStructs
from rdkit.Chem import AllChem
from sklearn.manifold import TSNE
```

```python
def getFP_by_data(smiles_list, marker = 0):
    # df = pd.read_csv('./data/drd2_active.csv')  
    # smiles_list = list(set(smiles_list))

    MorganFP = []
    for smiles in smiles_list:
        m = Chem.MolFromSmiles(smiles)
        # fp = AllChem.GetMorganFingerprintAsBitVect(m, 3, 2048).ToList()
        fp = AllChem.GetMorganFingerprintAsBitVect(m, 3, 2048)
        MorganFP.append(fp)

    # MorganFP = np.array(MorganFP)
    labels = [marker for i in range(len(MorganFP))]
    labels = np.array(labels)

    return smiles_list,MorganFP, labels

def get_list_fps(fps):
    fps_list = []
    for fp in fps:
        fps_list.append(fp.ToList())
    fps_list = np.array(fps_list)
    return fps_list

def plot_embedding_2d(X, labels, bp_list, legends, title=None, fname='None'):
    """Plot an embedding X with the class label y colored by the domain d."""
    x_min, x_max = np.min(X, 0), np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)

    # Plot colors numbers
    # num_of_labels = labels.max().item() + 1

    plt.figure(figsize=(16,10))
    # ax = plt.subplot(111)

    # colors_space = np.linspace(0, 1, num_of_labels)           # 生成颜色空间
    # label_to_color = {}                                 # 将标签对应为颜色
    # for i in range(num_of_labels):
    #     label_to_color[i] = colors_space[i]

    s = 0
    e = 0
    for i in range(len(bp_list)):
        bp = bp_list[i]
        s = e
        e = e + bp

        colors = labels[s:e]
        lb = legends[i]

        # print('==============> ', lb, s ,e)
        # print(colors[:10])
        sc = plt.scatter(X[s:e, 0], X[s:e, 1], c=colors, s=20, label=lb)


    # colors = []
    # for label in labels:
    #     colors.append(label)

    # sc = plt.scatter(X[:, 0], X[:, 1], c=colors, s=20)
    # scatter = mscatter(X[:, 0], X[:, 1], c='r', ax=ax)

    plt.xticks([]), plt.yticks([])
    if title is not None:
        plt.title(title)
    # plt.colorbar(sc)

    plt.legend()
    plt.savefig(fname, bbox_inches='tight')
    plt.show()
```

```python
# fname = './outputs/bbb_20231003172239.csv'
# fname = './outputs/bbb_qed_20231003183427.csv'
# fname = './outputs/bbb_qed_sa_20231003202759.csv'
# fname = './outputs/hia_20231003211638.csv'
# fname = './outputs/hia_qed_20231004125027.csv'
# fname = './outputs/hia_qed_sa_20231004141910.csv'

fps_list = []
labels_list = []
fname_list = [
                ('./outputs/drd2_0_50_mt.csv','DRD2'),
                ('./outputs/drd2_qed_0_10_mt.csv','DRD2+QED'),
                # ('./outputs/drd2_qed_sa_0_10_mt.csv','DRD2+QED+SAs'),
                ('./outputs/gsk3_0_10_mt.csv','GSK3'),
                ('./outputs/gsk3_qed_0_10_mt.csv','GSK3+QED'),
                # ('./outputs/gsk3_qed_sa_0_10_mt.csv','GSK3+QED+SAs'),
                # ('./outputs/bbb_20231003172239_mt.csv','BBB'),
                ('./outputs/bbb_qed_20231003183427_mt.csv','BBB+QED'),
                # ('./outputs/bbb_qed_sa_20231003202759_mt.csv','BBB+QED+SAs'),
                # ('./outputs/hia_20231003211638_mt.csv','HIA'),
                ('./outputs/hia_qed_20231004125027_mt.csv','HIA+QED'),
                # ('./outputs/hia_qed_sa_20231004141910_mt.csv','HIA+QED+SAs')
                ]
```

```python
for i in range(len(fname_list)):
    fname = fname_list[i][0]
    df = pd.read_csv(fname)
    mask = df['IsSuccess'] == True
    smiles_list = df['smiles'][mask][:1000]
    _, fps, labels = getFP_by_data(smiles_list.values, f'C{i}')
    fps = get_list_fps(fps)
    fps_list.append(fps)
    labels_list.append(labels)
```

```python
bp_list = [len(i) for i in fps_list]
bp_list
```

```python
legends = [item[1] for item in fname_list]
```

```python
x = np.concatenate(fps_list)
y = np.concatenate(labels_list)
```

```python
tsne2d = TSNE(n_components=2, init='pca', random_state=0)
X_tsne_2d = tsne2d.fit_transform(x)
# plot_embedding_2d(X_tsne_2d, y, bp_list, legends, title = "Molecular distribution under different tasks")
plot_embedding_2d(X_tsne_2d, y, bp_list, legends)
```
